```python
# Function definitions

def CustomerInput():             # Store input for items/quantities and customer details, in CustInp dictionary

    global CustInp
    CustInp= {}

    for item in stock.keys():         # Store customer input for Qty required, also check if entry is valid
        ItemDetails = stock.get(item)
        Item = ItemDetails[3]
        valid = 0
        while valid == 0:
            #print('\n',Item,':',ItemDetails[1],'packets available: ',end=' ')
            print('\n',Item,':',end=' ')
            Qty = input('How many required?    ')

            if int(ItemDetails[1]) == 0:            # Check if items are out of stock, exit while loop
                print(Item,'Out of stock!!\n')
                Qty = int(ItemDetails[1])
                CustInp.setdefault(item,int(Qty))   # Store customer input for items/quantities
                valid = 1

            elif Qty.strip() == '' or Qty.isdigit() == False:  valid = 0   # Qty invalid, stay in while loop
            elif int(Qty) <= int(ItemDetails[1]):   # Qty valid, exit while loop
                valid = 1
                CustInp.setdefault(item,int(Qty))   # Store customer input for items/quantities
            elif int(ItemDetails[1]) != 0 and int(Qty) > int(ItemDetails[1]):  valid = 0      # Qty invalid, stay in while loop

    DeliveryDetails = ['name','address']         # Store input for customer details, also check if entry is valid
    print('\n\nFill in your delivery details: ')
    for d in DeliveryDetails:
        ValidDelDet = 0
        while ValidDelDet == 0:                  # while loop to ask the user for input again, if blank or only number is entered.
            print("\nEnter your",d,":    ",end=' ')
            inp = input()
            if inp == '' or inp.isnumeric():
                print('Invalid',d,'entered\n')
                ValidDelDet=0
            else : ValidDelDet=1
            CustInp.setdefault(d,inp.title())

    Distance='x'
    while Distance.upper() not in ['A','B','C','D']:
        print('\nEnter distance of delivery address from store as (A\B\C\D)- \nA: 0-5 km | B: 5-10 km | C: 10-15 km | D: Above 15 km:  ',end=' ')
        Distance = input()
        Distance = Distance.strip()
    CustInp.setdefault('Distance',Distance.upper())        # Store customer input for delivery distance from store,in CustInp dictionary


def CalcBill():

    global TotalCost, TotalItemCost, TotalBillAmt, DeliveryCharges
    TotalCost= {}
    TotalItemCost = 0
    for item in stock.keys():                            # calculate and store TotalItemCost, DeliveryCharges and TotalBillAmt.
        CustDetails = CustInp.get(item)
        ItemDetails = stock.get(item)

        if int(ItemDetails[1]) < CustInp.get(item):      # Check if items are in stock
            TotalCost.setdefault(item,'0.0')             # No amt to be paid, if out of stock
        else:                                            # Else calculate and store final cost.                                               # calculate and store final cost.
            cost = ItemDetails[2]
            Tc = float(cost)*CustDetails
            TotalCost.setdefault(item,Tc)
            TotalItemCost= round((TotalItemCost+Tc),2)

    if TotalItemCost == 0: DeliveryCharges = 0
    elif CustInp.get('Distance') == 'A': DeliveryCharges = 10     # Calculate delivery charges based on user input.
    elif CustInp.get('Distance') == 'B': DeliveryCharges = 12     # A: 0-5  B: 5-10  C: 10-15  D: Above 15
    elif CustInp.get('Distance') == 'C': DeliveryCharges = 14
    elif CustInp.get('Distance') == 'D': DeliveryCharges = 20

    TotalBillAmt = DeliveryCharges + TotalItemCost


def Display(Display):

    if Display == 'DisplayStock':        # Display available stock before the customer

        print('\n\n\t\t\t\t\t\t ESSENTIAL ITEMS AVAILABLE\n')
        print('\t\tS.No\t\t\tItem\t\t\t\tQuantity\t\tCost/Item ($)')

        for item in stock.keys():
            ItemDetails = stock.get(item)
            print('\t\t',item,'\t\t', ItemDetails[0],'\t\t  ', ItemDetails[1], '\t\t\t ', ItemDetails[2])
        print("\n")

    elif Display == 'DisplayBill':       # Display final bill with customer details
        print("\n\n")
        print(('*') * 125)
        print('\n\t\t\t\t\t\t\t BILL\n')
        print('\t\tS.No\t\t\tItem\t\t\t\tQuantity\t\tTotal Cost ($)')

        for inp in stock.keys():
            CustDetails = CustInp.get(inp)
            ItemDetails = stock.get(inp)
            Cost = TotalCost.get(inp)
            print('\t\t',inp,'\t\t', ItemDetails[0],'\t\t  ', CustDetails, '\t\t\t ', Cost)

        print('\nTotal Item Cost   = $',TotalItemCost,sep = '')
        print('\nDelivery Charges  = $',DeliveryCharges,sep = '')
        print('\nTotal Bill Amount = $',TotalBillAmt,sep = '')
        print('(Total Item Cost + Delivery Charges)')

        print('\nName:   ',CustInp['name'])
        print('Address:',CustInp['address'])
        print('\n',(' ') * 50,'Thanks for shopping!\n')
        print(('*') * 125)
        print("\n\n")


def UpdateStock():        # Update the quantities of remaining items in stock, before next transaction

    global stock
    for item in stock.keys():
        CustDetails = CustInp.get(item)
        ItemDetails = stock.get(item)

        if int(CustInp.get(item)) <= int(ItemDetails[1]):
            ItemDetails[1] = int(ItemDetails[1]) - int(CustInp.get(item))
            stock.setdefault(item,ItemDetails[1])                          # updating stock dictionary based on CustInp dictionary
```

```python
#Main Program

stock= {'1':['Biscuits: a pack of 20','5','2.50', 'Biscuits'],'2':['Cereals: 2 pounds per pack','10','3.50', 'Cereals'],
        '3':['Chicken: 1 pounds per pack','20','13.30' ,'Chicken'],'4':['Oats: 4 pounds per pack','20','3.90', 'Oats'],
        '5':['Rice: 2 pounds per pack','30','4.50', 'Rice']}

Repeat='Y'
while (Repeat.upper() != 'N'):                   # loop to continue shopping, until user enters N
# functions calls
    Display('DisplayStock')        # to display Stock
    CustomerInput()                         # to take customer input
    CalcBill()                      # to calculate bill
    Display('DisplayBill')         # to display final bill

    Repeat = input("\n\nDo you want to start shopping? (Y/N): ")
    if Repeat.upper() == 'Y':
        UpdateStock()               # to update Stock , if user wants to continue shopping
        x=0
        for i in stock.values(): x= x+int(i[1])  # check if no stock available, terminate program.
        if x == 0:
            Repeat = 'N'
            print("ALL ITEMS OUT OF STOCK!!!")
print("TERMINATING..")    1
```
